```python
from path import Path
import pandas as pd
```

```python
# Load the ANANTH R dataset retrieved from Kaggle
ananthr1 = Path('../Room_Occupancy_Project/Resources/Kaggle_Room_Estimation/Dataset/Occupancy_Estimation.csv')
ananthr1_df = pd.read_csv(ananthr1, )
ananthr1_df.head()
```

```python
# Load the Sachin Sharma dataset retrieved from Kaggle
sharma = Path('../Room_Occupancy_Project/Resources/Kaggle_Room_Occupancy/Dataset/file.csv')
sharma_df = pd.read_csv(sharma)
sharma_df.head()
```

```python
# Load the kukuroo3 dataset retrieved from Kaggle.
kukuroo3 = Path('../Room_Occupancy_Project/Resources/Kaggle_Room_Occupancy_Detection/Dataset/Occupancy.csv')
kukuroo3_df = pd.read_csv(kukuroo3)
kukuroo3_df.head()
```

```python
# The ANANTH R dataset is not a binary classification.  Instead, it attempts to detect
# the number of people in a room (0, 1, 2, or 3).  The goal for this project is to simply determine if one
# or more people are in a room (binary classification).  Convert values 1, 2, and 3 to 1 to indicate
# occupied.  No change to rows with value 0.
ananthr1_df.loc[ananthr1_df['Room_Occupancy_Count'] > 0, ['Room_Occupancy_Count']] = 1
ananthr1_df = ananthr1_df.rename(columns={'Room_Occupancy_Count' : 'Occupancy'})
ananthr1_df.head()
```

```python
#change room occupancy count to occupancy in Data 1

ananthr1_df.rename(columns={'Room_Occupancy_Count':'Occupancy'}, inplace=True)
ananthr1_df.head()
```

```python
# Import database libraries for working with sqlite
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
```

```python
# Connect the database and load the tables.  The purpose of this workbook is to establish
# the baseline database, so if the tables already exist we can replace them with the entire 
# DataFrame
db_engine = create_engine("sqlite:///sensor-telemetry.sqlite")
kukuroo3_df.to_sql(name="kukuroo", con=db_engine, if_exists='replace')
sharma_df.to_sql(name="sharma", con=db_engine, if_exists='replace')
ananthr1_df.to_sql(name="ananth", con=db_engine, if_exists='replace')
```
